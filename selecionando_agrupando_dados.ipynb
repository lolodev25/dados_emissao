#vai juntas as colunas e calcular a diferença de emissão de gases por ano

```python
#juntando as colunas
colunas_info = df_emissoes_geral.loc[:, 'Nível 1 - Setor':'Produto'].columns
colunas_info
```

```python
#juntando as de anos

coluna_anos = df_emissoes_geral.loc[:,1970:2021].columns
coluna_anos
```

```python
emissoes_po_ano = df_emissoes_geral.melt(id_vars = colunas_info, value_vars = coluna_anos, var_name='Ano', value_name = 'Emissão')
emissoes_po_ano
```

```python
#agrupando os tipos de gases
emissoes_po_ano.groupby('Gás').groups
```

```python
#vendo a emissão de gases de um grupo específico
emissoes_po_ano.groupby('Gás').get_group('CO2 (t)')
```

```python
#soma de emisão para cada tipo de gás
emissoes_po_ano.groupby('Gás')[['Emissão']].sum()
```

```python
emisssao_por_gas = emissoes_po_ano.groupby('Gás')[['Emissão']].sum().sort_values('Emissão', ascending= False).reset_index()
emisssao_por_gas
```

```python
emisssao_por_gas.plot(x='Gás', y='Emissão', kind='bar', title='Emissões de gáses')
```

```python
emisssao_por_gas.iloc[0:9]
```

```python
emissao_co2 = (emisssao_por_gas.iloc[0:9])
soma_co2 = (emisssao_por_gas.iloc[0:9].sum())

total = soma_co2['Emissão'].sum()/emisssao_por_gas['Emissão'].sum()
total
```

```python
print(f'A emissão de C03 corresponde a {float((total))*100:.2f}')
```

#emissão de gases por setor

```python
gas_por_setor = emissoes_po_ano.groupby(['Gás', 'Nível 1 - Setor'])[['Emissão']].sum()
gas_por_setor
```

```python
gas_por_setor.xs('CO2 (t)', level = 0)
#imprime somente os dados do gas CO2
```

```python
#mostrando o CO2 e a mudança de uso da terra e floresta
gas_por_setor.xs(('CO2 (t)','Mudança de Uso da Terra e Floresta'), level = [0,1])
```

```python
#vendo qual e o com emissão maxima, vai mostrar o nome da linha
gas_por_setor.xs('CO2 (t)', level = 0).idxmax()
```

```python
#mostrando os gases e qual atividade economica emite mais
gas_por_setor.groupby(level = 0).idxmax()
```

```python
gas_por_setor.groupby(level = 0).max()
```

```python
#descantarndo o index
valores_max = gas_por_setor.groupby(level = 0).max().values
```

```python
tabela_sumarizada = gas_por_setor.groupby(level = 0).idxmax()
tabela_sumarizada.insert(1, 'Quantidade de emissão',valores_max)
tabela_sumarizada
```

```python
gas_por_setor.swaplevel(0,1).groupby(level = 0).idxmax()
```
